2. ‡∏ù‡∏∂‡∏Å‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô Pattern ‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢
3. ‡∏•‡∏≠‡∏á‡∏•‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏¢‡∏∞ (Garbage Data) ‡∏à‡∏£‡∏¥‡∏á‡πÜ





### Basic Syntax (‡∏™‡∏±‡∏ç‡∏•‡∏±‡∏Å‡∏©‡∏ì‡πå‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô)

#### 1. Character Classes (‡∏ï‡∏±‡∏ß‡πÅ‡∏ó‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£)

*   . : ‡πÅ‡∏ó‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡πÑ‡∏î‡πâ
*   \d : ‡πÅ‡∏ó‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
*   \s : ‡πÅ‡∏ó‡∏ô‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á
*   [] : ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ï‡∏±‡∏ß‡πÉ‡∏î‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡∏∂‡πà‡∏á


```python
import re

# ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ä‡πà‡∏ß‡∏¢‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• (Helper Function)
def show_match(pattern, text, desc):
    matches = re.findall(pattern, text)
    print(f"üìå ‡πÇ‡∏à‡∏ó‡∏¢‡πå: {desc}")
    print(f"   Pattern: {pattern}")
    print(f"   ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°: '{text}'")
    print(f"   ‚úÖ ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏à‡∏≠: {matches}")
    print("-" * 50)
```

```python
# \d : ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
show_match(r"\d", "‡∏â‡∏±‡∏ô‡∏™‡∏π‡∏á 180 ‡∏´‡∏ô‡∏±‡∏Å 75", r"‡∏´‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÇ‡∏î‡∏î‡πÜ (\d)")

# \w : ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£, ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç, _ (Word Character)
# ‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏: ‡πÉ‡∏ô Python 3 \w ‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢‡∏î‡πâ‡∏ß‡∏¢
show_match(r"\w+", "ID: User_01 ‡∏ä‡∏∑‡πà‡∏≠: ‡∏™‡∏°‡∏ä‡∏≤‡∏¢", r"‡∏´‡∏≤‡∏Ñ‡∏≥‡∏´‡∏£‡∏∑‡∏≠‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ (\w+)")

# \s : ‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á (Space, Tab, Enter)
show_match(r"‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ\s‡∏Ñ‡∏£‡∏±‡∏ö", "‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ ‡∏Ñ‡∏£‡∏±‡∏ö", r"‡∏´‡∏≤‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡∏°‡∏µ‡πÄ‡∏ß‡πâ‡∏ô‡∏ß‡∏£‡∏£‡∏Ñ‡∏Ñ‡∏±‡πà‡∏ô (\s)")

# . : ‡∏ï‡∏±‡∏ß‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡πÑ‡∏î‡πâ 1 ‡∏ï‡∏±‡∏ß (Wildcard)
show_match(r"‡∏Å.‡∏£", "‡∏Å‡∏≤‡∏£ ‡∏Å‡∏£ ‡∏Å‡∏π‡∏£ ‡∏Å-‡∏£", r"‡∏Å ‡πÑ‡∏Å‡πà + ‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡πÑ‡∏î‡πâ + ‡∏£ ‡πÄ‡∏£‡∏∑‡∏≠ (.)")
```

#### 2. ‡∏Å‡∏≤‡∏£‡∏£‡∏∞‡∏ö‡∏∏‡∏à‡∏≥‡∏ô‡∏ß‡∏ô (Quantifiers)
‡πÉ‡∏ä‡πâ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Å‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡∏ã‡πâ‡∏≥‡πÜ ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÅ‡∏ô‡πà‡∏ô‡∏≠‡∏ô

*   "*" : ‡∏°‡∏µ‡∏Å‡∏µ‡πà‡∏ï‡∏±‡∏ß‡∏Å‡πá‡πÑ‡∏î‡πâ (0 ‡∏´‡∏£‡∏∑‡∏≠ ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤)
*   "+" : ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß
*   {n} : ‡∏£‡∏∞‡∏ö‡∏∏‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏õ‡πä‡∏∞‡πÜ










```python
# --- 2. ‡∏ï‡∏±‡∏ß‡∏£‡∏∞‡∏ö‡∏∏‡∏à‡∏≥‡∏ô‡∏ß‡∏ô (Quantifiers) ---

# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏î‡∏±‡∏Å‡∏à‡∏±‡∏ö‡πÄ‡∏™‡∏µ‡∏¢‡∏á‡∏´‡∏±‡∏ß‡πÄ‡∏£‡∏≤‡∏∞ (555+)
text_laugh = "‡∏ï‡∏•‡∏Å‡∏°‡∏≤‡∏Å 55555 ‡∏´‡∏£‡∏∑‡∏≠‡πÅ‡∏Ñ‡πà 5 ‡∏ï‡∏±‡∏ß‡πÄ‡∏î‡∏µ‡∏¢‡∏ß"
show_match(r"5+", text_laugh, r"‡∏´‡∏≤‡πÄ‡∏•‡∏Ç 5 ‡∏ó‡∏µ‡πà‡∏°‡∏≤‡∏ï‡∏¥‡∏î‡∏Å‡∏±‡∏ô (5+)")

# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏î‡∏±‡∏Å‡∏à‡∏±‡∏ö‡∏Ñ‡∏≥‡∏•‡∏≤‡∏Å‡πÄ‡∏™‡∏µ‡∏¢‡∏á‡∏¢‡∏≤‡∏ß (‡πÄ‡∏ä‡πà‡∏ô ‡∏°‡∏≤‡∏Å‡∏Å‡∏Å‡∏Å)
text_drag = "‡∏ä‡∏≠‡∏ö‡∏°‡∏≤‡∏Å‡∏Å‡∏Å‡∏Å‡∏Å‡∏Å ‡∏´‡∏£‡∏∑‡∏≠‡πÅ‡∏Ñ‡πà ‡∏ä‡∏≠‡∏ö‡∏°‡∏≤‡∏Å"
show_match(r"‡∏°‡∏≤‡∏Å+", text_drag, r"‡∏´‡∏≤‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ '‡∏°‡∏≤‡∏Å' ‡∏ó‡∏µ‡πà ‡∏Å.‡πÑ‡∏Å‡πà ‡∏¢‡∏≤‡∏ß‡πÜ")

# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏£‡∏´‡∏±‡∏™‡πÑ‡∏õ‡∏£‡∏©‡∏ì‡∏µ‡∏¢‡πå 5 ‡∏´‡∏•‡∏±‡∏Å
text_zip = "‡∏ö‡πâ‡∏≤‡∏ô‡∏ú‡∏°‡∏≠‡∏¢‡∏π‡πà‡πÄ‡∏Ç‡∏ï 10400 ‡∏´‡∏£‡∏∑‡∏≠ 10500 ‡∏ô‡∏∞"
show_match(r"\d{5}", text_zip, r"‡∏´‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 5 ‡∏´‡∏•‡∏±‡∏Å‡πÄ‡∏õ‡πä‡∏∞‡πÜ (\d{5})")
```

#### 3. Anchors (‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á)


*   ^ : ‡∏ï‡πâ‡∏≠‡∏á‡∏Ç‡∏∂‡πâ‡∏ô‡∏ï‡πâ‡∏ô‡∏î‡πâ‡∏ß‡∏¢...
*   $ : ‡∏ï‡πâ‡∏≠‡∏á‡∏•‡∏á‡∏ó‡πâ‡∏≤‡∏¢‡∏î‡πâ‡∏ß‡∏¢...



```python
# ^ : ‡∏ï‡πâ‡∏≠‡∏á‡∏Ç‡∏∂‡πâ‡∏ô‡∏ï‡πâ‡∏ô‡∏î‡πâ‡∏ß‡∏¢...
# ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï: ‡πÄ‡∏à‡∏≠ '‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ' ‡πÅ‡∏Ñ‡πà‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡πâ‡∏≤‡∏™‡∏∏‡∏î ‡∏ï‡∏±‡∏ß‡∏´‡∏•‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤
show_match(r"^‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ", "‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ‡∏Ñ‡∏£‡∏±‡∏ö... ‡∏≠‡πâ‡∏≤‡∏ß ‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ", r"‡∏ï‡πâ‡∏≠‡∏á‡∏Ç‡∏∂‡πâ‡∏ô‡∏ï‡πâ‡∏ô‡∏õ‡∏£‡∏∞‡πÇ‡∏¢‡∏Ñ‡∏î‡πâ‡∏ß‡∏¢ ‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ (^)")

# $ : ‡∏ï‡πâ‡∏≠‡∏á‡∏•‡∏á‡∏ó‡πâ‡∏≤‡∏¢‡∏î‡πâ‡∏ß‡∏¢...
# ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï: ‡πÄ‡∏à‡∏≠ '‡∏Ñ‡∏£‡∏±‡∏ö' ‡πÅ‡∏Ñ‡πà‡∏ï‡∏±‡∏ß‡∏ó‡πâ‡∏≤‡∏¢‡∏™‡∏∏‡∏î ‡∏ï‡∏±‡∏ß‡∏Å‡∏•‡∏≤‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤
show_match(r"‡∏Ñ‡∏£‡∏±‡∏ö$", "‡∏Ñ‡∏£‡∏±‡∏ö‡∏ú‡∏°... ‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡∏£‡∏±‡∏ö", r"‡∏ï‡πâ‡∏≠‡∏á‡∏•‡∏á‡∏ó‡πâ‡∏≤‡∏¢‡∏õ‡∏£‡∏∞‡πÇ‡∏¢‡∏Ñ‡∏î‡πâ‡∏ß‡∏¢ ‡∏Ñ‡∏£‡∏±‡∏ö ($)")
```

#### 4. Groups & Sets (‡∏Å‡∏•‡∏∏‡πà‡∏°‡πÅ‡∏•‡∏∞‡∏ó‡∏≤‡∏á‡πÄ‡∏•‡∏∑‡∏≠‡∏Å)


*   [...] : ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ï‡∏±‡∏ß‡πÉ‡∏î‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡∏∂‡πà‡∏á‡πÉ‡∏ô‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö (Set)
*   [^...] : ‡∏´‡πâ‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏´‡∏•‡πà‡∏≤‡∏ô‡∏µ‡πâ (Negation)

*   | : ‡∏´‡∏£‡∏∑‡∏≠ (Or)
*   (...) : ‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏° (Group)





```python
# [...] : ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ï‡∏±‡∏ß‡πÉ‡∏î‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡∏∂‡πà‡∏á‡πÉ‡∏ô‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö (Set)
show_match(r"‡∏ó[‡∏≥‡∏°]", "‡∏ó‡∏≥ ‡∏ó‡∏° ‡∏ó‡∏≤ ‡∏ó‡∏¢", r"‡∏ó ‡∏ï‡∏≤‡∏°‡∏î‡πâ‡∏ß‡∏¢ ‡∏™‡∏£‡∏∞‡∏≠‡∏≥ ‡∏´‡∏£‡∏∑‡∏≠ ‡∏° ‡∏°‡πâ‡∏≤ [‡∏≥‡∏°]")

# [^...] : ‡∏´‡πâ‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏´‡∏•‡πà‡∏≤‡∏ô‡∏µ‡πâ (Negation)
# ‡πÅ‡∏õ‡∏•‡∏ß‡πà‡∏≤: ‡πÄ‡∏≠‡∏≤‡∏ó‡∏∏‡∏Å‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
show_match(r"[^0-9]+", "‡∏õ‡∏µ 2024 Year 2567", r"‡∏´‡∏≤‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡πÑ‡∏î‡πâ‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç ([^0-9])")

# | : ‡∏´‡∏£‡∏∑‡∏≠ (Or)
show_match(r"‡πÅ‡∏°‡∏ß|‡∏´‡∏°‡∏≤", "‡∏â‡∏±‡∏ô‡∏£‡∏±‡∏Å‡πÅ‡∏°‡∏ß‡πÅ‡∏•‡∏∞‡∏´‡∏°‡∏≤", r"‡πÅ‡∏°‡∏ß ‡∏´‡∏£‡∏∑‡∏≠ ‡∏´‡∏°‡∏≤ (|)")

# (...) : ‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏° (Group)
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏£‡∏∞‡∏ö‡∏∏‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ç‡∏≠‡∏á "‡∏ó‡∏±‡πâ‡∏á‡∏Ñ‡∏≥" ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡πÅ‡∏Ñ‡πà‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÄ‡∏î‡∏µ‡∏¢‡∏ß
show_match(r"(555)+", "‡∏ï‡∏•‡∏Å‡∏à‡∏±‡∏á 555555", r"‡∏´‡∏≤‡∏ä‡∏∏‡∏î‡∏´‡∏±‡∏ß‡πÄ‡∏£‡∏≤‡∏∞ 555 ‡∏ã‡πâ‡∏≥‡πÜ (...)+")
```

```python
import re

text = "‡∏ï‡∏¥‡∏î‡∏ï‡πà‡∏≠‡∏Ñ‡∏∏‡∏ì A ‡πÇ‡∏ó‡∏£ 081-111-1111 ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡∏∏‡∏ì B ‡πÇ‡∏ó‡∏£ 082-222-2222 ‡∏î‡πà‡∏ß‡∏ô"
pattern_phone = r"\d{3}-\d{3}-\d{4}" # ‡πÅ‡∏û‡∏ó‡πÄ‡∏ó‡∏¥‡∏£‡πå‡∏ô‡πÄ‡∏ö‡∏≠‡∏£‡πå‡πÇ‡∏ó‡∏£

print(f"‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö: {text}\n")

# ---------------------------------------------------------
# 1. re.findall() -> ‡∏û‡∏£‡∏∞‡πÄ‡∏≠‡∏Å‡∏á‡∏≤‡∏ô Extraction (‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•)
# ---------------------------------------------------------
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠: ‡∏≠‡∏¢‡∏≤‡∏Å‡πÑ‡∏î‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• "‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î" ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏ï‡∏≤‡∏°‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç
phones = re.findall(pattern_phone, text)
print(f"1. findall (‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•): {phones}")
# ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå: ['081-111-1111', '082-222-2222']


# ---------------------------------------------------------
# 2. re.sub() -> ‡∏û‡∏£‡∏∞‡πÄ‡∏≠‡∏Å‡∏á‡∏≤‡∏ô Cleaning (‡∏•‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•)
# ---------------------------------------------------------
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠: ‡∏≠‡∏¢‡∏≤‡∏Å‡∏•‡∏ö ‡∏´‡∏£‡∏∑‡∏≠ ‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà‡∏Ñ‡∏≥ (Data Filtering ‡πÉ‡∏ä‡πâ‡∏≠‡∏±‡∏ô‡∏ô‡∏µ‡πâ‡πÄ‡∏¢‡∏≠‡∏∞‡∏™‡∏∏‡∏î!)
# (Pattern, ‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡∏à‡∏∞‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà, ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö)
censored_text = re.sub(pattern_phone, "[XXX]", text)
print(f"2. sub (‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà/‡∏•‡∏ö):    {censored_text}")
# ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå: ‡∏ï‡∏¥‡∏î‡∏ï‡πà‡∏≠‡∏Ñ‡∏∏‡∏ì A ‡πÇ‡∏ó‡∏£ [XXX] ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡∏∏‡∏ì B ‡πÇ‡∏ó‡∏£ [XXX] ‡∏î‡πà‡∏ß‡∏ô


# ---------------------------------------------------------
# 3. re.search() -> ‡∏û‡∏£‡∏∞‡πÄ‡∏≠‡∏Å‡∏á‡∏≤‡∏ô Validation (‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö)
# ---------------------------------------------------------
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠: ‡∏≠‡∏¢‡∏≤‡∏Å‡∏£‡∏π‡πâ‡πÅ‡∏Ñ‡πà‡∏ß‡πà‡∏≤ "‡∏°‡∏µ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà‡∏°‡∏µ" (‡πÄ‡∏≠‡∏≤‡πÑ‡∏õ‡πÉ‡∏ä‡πâ‡πÉ‡∏ô if-else)
match = re.search(r"‡∏î‡πà‡∏ß‡∏ô", text)
if match:
    print(f"3. search (‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö):  ‡πÄ‡∏à‡∏≠‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤‡∏î‡πà‡∏ß‡∏ô! ‡∏ó‡∏µ‡πà‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á {match.span()}")
else:
    print(f"3. search (‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö):  ‡πÑ‡∏°‡πà‡πÄ‡∏à‡∏≠")


# ---------------------------------------------------------
# 4. re.split() -> ‡∏ï‡∏±‡∏ß‡∏ä‡πà‡∏ß‡∏¢‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥
# ---------------------------------------------------------
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠: ‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡πà‡∏ô (Delimiter) ‡πÑ‡∏°‡πà‡πÅ‡∏ô‡πà‡∏ô‡∏≠‡∏ô (‡πÄ‡∏ä‡πà‡∏ô ‡∏ö‡∏≤‡∏á‡∏ó‡∏µ‡πÉ‡∏ä‡πâ comma, ‡∏ö‡∏≤‡∏á‡∏ó‡∏µ‡πÉ‡∏ä‡πâ space)
messy_string = "apple,banana;orange lemon"
# ‡∏ï‡∏±‡∏î‡∏î‡πâ‡∏ß‡∏¢ , ‡∏´‡∏£‡∏∑‡∏≠ ; ‡∏´‡∏£‡∏∑‡∏≠ ‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á
fruits = re.split(r"[,;\s]+", messy_string)
print(f"4. split (‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥):     {fruits}")
# ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå: ['apple', 'banana', 'orange', 'lemon']
```

```python
def clean_data_pipeline(text):
    print(f"üî¥ ORIGINAL:\n{text}\n")

    # 1. ‡∏•‡∏ö HTML Tags (‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡∏ï‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô <...>)
    text = re.sub(r"<[^>]+>", "", text)

    # 2. ‡∏•‡∏ö URL (http/https ‡∏´‡∏£‡∏∑‡∏≠ www)
    text = re.sub(r"https?://\S+|www\.\S+", "[URL_REMOVED]", text)

    # 3. Masking ‡πÄ‡∏ö‡∏≠‡∏£‡πå‡πÇ‡∏ó‡∏£‡∏®‡∏±‡∏û‡∏ó‡πå (Privacy - PDPA ‡πÄ‡∏ö‡∏∑‡πâ‡∏≠‡∏á‡∏ï‡πâ‡∏ô)
    # ‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö: 0 ‡∏ï‡∏≤‡∏°‡∏î‡πâ‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç 9 ‡∏´‡∏•‡∏±‡∏Å (‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢)
    text = re.sub(r"0\d{9}", "[PHONE]", text)

    # 4. ‡∏¢‡∏∏‡∏ö‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏ã‡πâ‡∏≥‡∏ã‡πâ‡∏≠‡∏ô (Multiple Spaces)
    text = re.sub(r"\s+", " ", text).strip()

    print(f"üü¢ CLEANED:\n{text}")
    return text

# --- ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡∏¥‡∏ö (‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏à‡∏≤‡∏Å Web Scraping) ---
dirty_text = """
<div>
   <h3>‡∏Ç‡πà‡∏≤‡∏ß‡∏î‡πà‡∏ß‡∏ô!</h3>
   ‡πÅ‡∏à‡∏Å‡πÄ‡∏Ñ‡∏£‡∏î‡∏¥‡∏ï‡∏ü‡∏£‡∏µ    ‡∏Ñ‡∏•‡∏¥‡∏Å‡πÄ‡∏•‡∏¢ https://bit.ly/game888

   ‡∏ï‡∏¥‡∏î‡∏ï‡πà‡∏≠‡∏Ñ‡∏∏‡∏ì‡∏™‡∏°‡∏ä‡∏≤‡∏¢ ‡πÇ‡∏ó‡∏£ 0812345678 ‡∏î‡πà‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡∏ö!!!
   <br>
   (‡∏´‡πâ‡∏≤‡∏°‡∏û‡∏•‡∏≤‡∏î)
</div>
"""

# ‡∏£‡∏±‡∏ô Pipeline
clean_data_pipeline(dirty_text)
```
